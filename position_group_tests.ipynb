# Defense/Special Teams

```python
week = 8
year = 2023
opp_team = "CIN"

rushing_yards_allowed = 0
passing_yards_allowed = 0
tds_allowed = 0

interceptions = 0
sacks = 0
sack_yards = 0
sack_fumbles = 0
sack_fumbles_recovered = 0
receiving_fumbles = 0
receiving_fumbles_recovered = 0

defense_game = data_2023[data_2023["season"].isin([year])]
defense_game = defense_game[defense_game["week"].isin([week])]
defense_game = defense_game[defense_game["recent_team"].isin([opp_team])]

rushing_yards_allowed = defense_game[["rushing_yards"]].sum()
print(rushing_yards_allowed)

passing_yards_allowed = defense_game[["passing_yards"]].sum()
print(passing_yards_allowed)

passing_tds_allowed = defense_game[["passing_tds"]].sum()
print(passing_tds_allowed)

rushing_tds_allowed = defense_game[["rushing_tds"]].sum()
print(rushing_tds_allowed)

special_teams_tds_allowed = defense_game[["special_teams_tds"]].sum()
print(special_teams_tds_allowed)

interceptions = defense_game[["interceptions"]].sum()
print(interceptions)

sacks = defense_game[["sacks"]].sum()
print(sacks)

sack_yards = defense_game[["sack_yards"]].sum()
print(sack_yards)

sack_fumbles = defense_game[["sack_fumbles"]].sum()
print(sack_fumbles)

sack_fumbles_recovered = defense_game[["sack_fumbles_lost"]].sum()
print(sack_fumbles_recovered)

receiving_fumbles = defense_game[["receiving_fumbles"]].sum()
print(receiving_fumbles)

receiving_fumbles_recovered = defense_game[["receiving_fumbles_lost"]].sum()
print(receiving_fumbles_recovered)

tds_allowed = int(passing_tds_allowed) + int(rushing_tds_allowed) + int(special_teams_tds_allowed)
print(tds_allowed)
```
